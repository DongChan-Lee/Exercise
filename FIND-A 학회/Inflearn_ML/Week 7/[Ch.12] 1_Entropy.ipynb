### Decision Tree
- Data를 가장 잘 구분할 수 있는 Tree를 구성하는 것이 목표!
- 어떤 질문(X, Attribute)이 가장 <b>많은 해답(Y)</b>을 줄 것인가?
- 어떤 질문이 <b>답의 모호성을 줄여줄 것인가?</b>
- Question을 통해 <b>splitting point</b>를 설정

### Entropy
- Entropy는 목적 달성을 위한 경우의 수를 정량적으로 표현하는 수치 → 작을수록 경우의 수가 적음
- Higher Entropy → Higher uncertainty
- Lower Entropy → Lower uncertainty
- <b>Entropy가 작으면 얻을 수 있는 정보가 명확하다!</b>
