### Youtube "3Blue1Brown"의 "Essence of calculus" 강의에 Gradient Descent 잘 정리되어 있음!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 11, 1)
f_x = x ** 2

plt.plot(x, f_x)
plt.xlim(-10.0, 10.0)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learning_rate = 0.01
for i in range(1000):
    old_value = x_new
    y.append(old_value ** 2)
    x_new = old_value - learning_rate * 2 * old_value
    derivative.append(x_new)
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

### Gradient descent에서 정해야 할 Parameter
- Learning rate에 대한 선정
- 얼마나 많이 loop을 돌 것인가?

#### Learning rate가 너무 작을 경우
- 끝까지 못감
- 시간이 오래 걸림


#### Learning rate가 너무 클 경우
- 데이터가 튀는 문제가 생김
- 수렴하지 못하는 경우 생김


### 굴곡이 많은 함수의 경우
- <strong>시작점에 따라 다른 최적값을 찾는다!</strong>
→ Stochastic Gradient Descent로 문제 해결 가능

- 식 넣어주면 자동으로 미분해주는 사이트
- http://www.wolframalpha.com/input/?i=derivative+x+sin(x%5E2)+%2B+1

```python
def sin_function(x):
    return x * np.sin(x ** 2) + 1

def derivative_f(x):
    return np.sin(x ** 2) + 2 * (x ** 2) * np.cos(x ** 2)
```

```python
x = np.arange(-3, 3.5, 0.001)
f_x = sin_function(x)

plt.plot(x, f_x)
plt.show()
```

```python
derivative_f(3)
```

```python
x_new = 1        # 이 시작점에 따라 다른 최적값을 찾음
derivative = []
y = []
learning_rate = 0.01
for i in range(10000):
    old_value = x_new
    y.append(sin_function(x_new))
    x_new = old_value - learning_rate * derivative_f(old_value)
    derivative.append(x_new)

plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```
