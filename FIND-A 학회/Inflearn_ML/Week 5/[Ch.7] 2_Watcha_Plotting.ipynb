```python
import matplotlib.pyplot as plt
import numpy as np
```

### Linear Regression with GD
- 결국 Parameter의 업데이트!!
- 임의의 θ0, θ1값으로 초기화
- Cost function J(θ0, θ1)이 최소화될 때까지 학습
- 더 이상 cost function이 줄어들지 않거나 학습 횟수를 초과할 때 종료

- Learning rate, Iteration 횟수 등 Parameter 지정
- Feature가 많으면 Normal equation에 비해 상대적으로 빠름
- 최적값에 수렴하지 않을 수도 있음

```python
x_data = np.array([8759, 10132, 12078, 16430])
y_data = np.array([487, 612, 866, 1030])

plt.scatter(x_data, y_data, color = 'r', marker = 's')
plt.show()
```

```python
weight_0 = np.random.sample(5)   # 0 ~ 1 사이 random value
weight_1 = np.random.sample(5)

weight_0, weight_1
```

```python
y_hat_list = []
for w0, w1 in zip(weight_0, weight_1):
    y_hat_list.append(w0 + w1 * x_data)   # hypothesis function 적용
y_hat_list
```

```python
plt.scatter(x_data, y_data, color = 'r', marker = 's')
for y_hat in y_hat_list:
    plt.plot(x_data, y_hat)
plt.show()
```

```python
y_hat[0] - y_data    # 예측값 - 실제값 : cost
```

```python
sum((y_hat - y_data) ** 2)
```
