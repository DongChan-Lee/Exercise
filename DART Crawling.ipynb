```python
def data_extract():
    url_lists = url_list
    data_finals = pd.DataFrame()
    for url in url_lists:
#         print(url)
        xmlsoup = BeautifulSoup(urlopen(url).read(), 'html.parser')
        te = xmlsoup.findAll("list")
        data = pd.DataFrame()
        for t in te:
            temp = pd.DataFrame(([[t.crp_cls.string, t.crp_nm.string, t.crp_cd.string, t.rpt_nm.string,
                              t.rcp_no.string, t.flr_nm.string, t.rcp_dt.string, t.rmk.string]]),
                              columns = ["crp_cls", "crp_nm", "crp_cd", "rpt_nm", "rcp_no", "flr_nm", "rcp_dt", "rmk"])
            data = pd.concat([data,temp])
            data_result = data
        data_result = data_result[data_result['rmk'] != '정연']
        data_finals = pd.concat([data_finals, data_result])
        data_finals = data_finals.reset_index(drop=True)
        data_finals = data_finals.drop_duplicates(['crp_cls', 'crp_nm', 'crp_cd', 'rpt_nm', 'rcp_no', 'flr_nm', 'rcp_dt', 'rmk'], keep='first')
#         print(data_finals)
    return data_finals
```

```python
# 입력받은 종목코드를 손질하고, 검색 중인 날짜 이전의 보고서만 나타나도록 
# 데이터프레임을 인덱싱하는 코드 추가 (date에 20191231 넣을 것임)
def get_rcp(code, date):
    from dateutil.parser import parse as p
    if len(code) == 7: code = code[1:]
    print(code)
    if type(date) == str: date = p(date)
    url = "http://dart.fss.or.kr/api/search.xml?auth=" + API_KEY + "&crp_cd=" + code + "&start_dt=20100101&end_dt=20191231&bsn_tp=A001&page_set=" + page_set
    xmlsoup = BeautifulSoup(urlopen(url).read(), 'html.parser')
    te = xmlsoup.findAll("list")
    data = pd.DataFrame()
    for t in te:
        temp = pd.DataFrame(([[t.crp_cls.string, t.crp_nm.string, t.crp_cd.string, t.rpt_nm.string,
                          t.rcp_no.string, t.flr_nm.string, t.rcp_dt.string, t.rmk.string]]),
                          columns = ["crp_cls", "crp_nm", "crp_cd", "rpt_nm", "rcp_no", "flr_nm", "rcp_dt", "rmk"])
        data = pd.concat([data,temp])
    data = data[data['rmk']!="정연"]
    data = data.reset_index(drop=True)
    return data
```
